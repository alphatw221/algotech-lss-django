```python
import django_initializer
from backend.pymongo.mongodb import db
from datetime import datetime
import arrow
```

```python
__collection = db.api_order
def get_anonymous_buyers_data(user_subscription_id):
    query = [
        {
            "$match":{
                "user_subscription_id": 1,
                "buyer":{'$eq':None}
            },
        },
        {
            '$project':{
                '_id':0,
                'id':1,
            }
        }
    ]
    cursor = __collection.aggregate(query)
    return list(cursor)

def get_registered_buyers_data(user_subscription_id):
    query = [
        {
            "$match":{
                "user_subscription_id": 1,
                "buyer_id":{'$ne':None}
            }
        },
        {
            '$group': {
                '_id': '$buyer_id',
                'id': {'$first':'$id'}
            }
        },
        {
            "$project": {
                '_id':0,
                'id': '$id'
            }
        }
    ]
    cursor = __collection.aggregate(query)
    return list(cursor)
```

```python
anonymous_buyers_data = get_anonymous_buyers_data(1)
registered_buyers_data = get_registered_buyers_data(1)

buyers_data = anonymous_buyers_data + registered_buyers_data
print(buyers_data)
```
